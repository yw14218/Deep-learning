```python
import pandas as pd
import matplotlib.pyplot as plt
data={'K':[0,1,2,3,4,5,6,7,8],
'loss':[0.62,0.70,0.73,0.72,0.76,0.77,0.78,0.80,0.80]}
valid= [0.60,0.65,0.67,0.65,0.68,0.66,0.67,0.66,0.65]
df = pd.DataFrame(data,columns=['K','loss'])
plt.plot(df['K'], df['loss'],label='train', color='red', marker='o')
plt.plot(df['K'], valid,label='valid', color='blue', marker='o')
plt.title('Effect of #of FC layers on network performance', fontsize=14)
plt.xlabel('K', fontsize=14)
plt.ylabel('Accuracy Score', fontsize=14)
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
N = 5
men_means = (0.61,0.86, 0.97, 0.98, 0.99)
women_means = (0.68,0.71, 0.77, 0.81, 0.80)
variance1 = [0.01,0.02, 0.01, 0.02, 0.01]
variance2 = [0.02,0.01, 0.01, 0.02, 0.05]
ind = np.arange(N) 
width = 0.35  
plt.bar(ind, men_means, width, label='Train',yerr=variance1)
plt.bar(ind + width, women_means, width,
    label='Valid',yerr=variance2)

plt.ylabel('Accuracy Score')
plt.title('Effect of #of filters in every layer on network performance')

plt.xticks(ind + width / 2, ('n=8','n=16', 'n=32', 'n=64', 'n=128'))
plt.legend(loc='best')
plt.grid(True)
plt.show()
```
